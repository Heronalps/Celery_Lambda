```python
import featuretools as ft
```

```python
data = ft.demo.load_mock_customer()
```

```python
customers_df = data["customers"]
sessions_df = data["sessions"]
transactions_df = data["transactions"]
```

```python
sessions_df
```

```python
transactions_df
```

```python
entities = {
    "customers" : (customers_df, "customer_id"),
    "sessions" : (sessions_df, "session_id", "session_start"),
    "transactions" : (transactions_df, "transaction_id", "transaction_time")
}
```

```python
entities
```

```python
# relationship (parent_entity, parent_variable, child_entity, child_variable)

relationships = [("sessions", "session_id", "transactions", "session_id"),
                 ("customers", "customer_id", "sessions", "customer_id")]
```

```python
feature_matrix_customers, feature_defs = ft.dfs(entities = entities, relationships = relationships, target_entity="customers")
```

```python
feature_matrix_sessions, feature_defs = ft.dfs(entities = entities, relationships = relationships, target_entity="sessions")
```

```python
transactions_df = data["transactions"].merge(data["sessions"]).merge(data["customers"])
```

```python
transactions_df.sample(10)
```

```python
products_df = data["products"]
```

```python
es = ft.EntitySet()
```

```python
es = es.entity_from_dataframe(entity_id="transactions", dataframe=transactions_df, index="transaction_id", time_index="transaction_time", variable_types={"product_id": ft.variable_types.Categorical})
```

```python
es
```

```python
es = es.entity_from_dataframe(entity_id="products", dataframe=products_df, index = "product_id")
```

```python
new_relationship = ft.Relationship(es["products"]["product_id"], es["transactions"]["product_id"])
```

```python
es = es.add_relationship(new_relationship)
```
