```python
import matplotlib.pyplot as plt
import pandas as pd
fontsize = 10
```

```python
'''
metric = ['Billed Duration', 'Max Memory Used']
'''

def metric_plot(model, metric):
    if metric == 'Billed Duration':
        unit = ' (ms)'
    elif metric == 'Max Memory Used':
        unit = ' (MB)'
    elif metric == 'execution_times':
        unit = ' (second)'
        
    df = pd.read_csv('../results/spreadsheets/{0}_{1}.csv'.format(model, metric))
    cols = [str(x) for x in list(range(1, 31))]
    df_metric = pd.DataFrame({'mean': df[cols].mean(), 'stdev' : df[cols].std()})
    # import pdb;pdb.set_trace();
    fig = plt.figure(figsize=[10, 10])
    ax = fig.add_subplot(111)
    ax.scatter(df_metric['mean'], df_metric['stdev'])
    # ax.set_xlim(300, 325)
    # ax.set_ylim(17, 22)
    ax.set_title( metric + unit, fontsize=fontsize)
    ax.set_ylabel("Standard Deviation", fontsize=fontsize)
    ax.set_xlabel("Mean", fontsize=fontsize)
    fig.savefig('../results/graphs/{0}_{1}.png'.format(model, metric))

def execution_time_plot(model, metric='execution_times'):
    df = pd.read_csv('../results/spreadsheets/{0}_{1}.csv'.format(model, metric))
    mean = df.loc[0].mean()
    stdev = df.loc[0].std()
    print ("{0} {1} mean : {2}".format(model, metric, mean))
    print ("{0} {1} stdev : {2}".format(model, metric, stdev))
```

```python
model = 'prophet'
metric = 'Billed Duration'
metric_plot(model, metric)

metric = 'Max Memory Used'  
metric_plot(model, metric)

metric = 'execution_times'  
execution_time_plot(model)
```
