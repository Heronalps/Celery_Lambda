```python
import matplotlib.pyplot as plt
import pandas as pd
fontsize = 15
```

```python
'''
metric = ['Billed Duration', 'Max Memory Used']
'''

def metric_plot(model, metric, metric_2):
    if metric == 'Billed Duration':
        unit = ' (ms)'
    elif metric == 'Max Memory Used':
        unit = ' (MB)'
    elif metric == 'execution_times':
        unit = ' (second)'
        
    df = pd.read_csv('../results/spreadsheets/{0}_{1}.csv'.format(model, metric))
    df_2 = pd.read_csv('../results/spreadsheets/{0}_{1}.csv'.format(model, metric_2))
    cols = [str(x) for x in list(range(1, 31))]
    df_metric = pd.DataFrame({'mean': df[cols].mean(), 'stdev' : df[cols].std()})
    df_metric_2 = pd.DataFrame({'mean': df_2[cols].mean(), 'stdev' : df_2[cols].std()})
    
    fig = plt.figure(figsize=[15, 30])
    ax = fig.add_subplot(511)
    ax.scatter(df_metric['mean'], df_metric['stdev'])
    # ax.set_xlim(300, 325)
    # ax.set_ylim(17, 22)
    ax.set_title( metric + unit, fontsize=fontsize)
    ax.set_ylabel("Standard Deviation", fontsize=fontsize)
    ax.set_xlabel("Mean", fontsize=fontsize)
    
    ax = fig.add_subplot(512)
    ax.set_title("Mean", fontsize=fontsize)
    ax.plot(df_metric['mean'])
    ax.set_ylabel(metric + "Mean", fontsize=fontsize)
    # ax2 = ax.twinx()
    ax = fig.add_subplot(513)
    ax.set_ylabel(metric_2 + "Mean", fontsize=fontsize)
    ax.set_xlabel("Trials", fontsize=fontsize)
    ax.plot(df_metric_2['mean'], color = 'r')
    
    ax = fig.add_subplot(514)
    ax.plot(df_metric['stdev'])
    ax.set_ylabel("Standard Deviation", fontsize=fontsize)
    ax.set_title("Standard Deviation", fontsize=fontsize)
    # ax2 = ax.twinx()
    ax = fig.add_subplot(515)
    ax.set_ylabel(metric_2 + "Mean", fontsize=fontsize)
    ax.set_xlabel("Trials", fontsize=fontsize)
    ax.plot(df_metric_2['stdev'], color = 'r')
        
    fig.savefig('../results/graphs/{0}_{1}.png'.format(model, metric))

def execution_time_plot(model, metric='execution_times'):
    df = pd.read_csv('../results/spreadsheets/{0}_{1}.csv'.format(model, metric))
    mean = df.loc[0].mean()
    stdev = df.loc[0].std()
    print ("{0} {1} mean : {2}".format(model, metric, mean))
    print ("{0} {1} stdev : {2}".format(model, metric, stdev))
    
def time_series_plot(model, metric, index):
    if metric == 'Billed Duration':
        unit = ' (ms)'
    elif metric == 'Max Memory Used':
        unit = ' (MB)'
    elif metric == 'execution_times':
        unit = ' (second)'
    df = pd.read_csv('../results/spreadsheets/{0}_{1}.csv'.format(model, metric))
    cols = [str(x) for x in list(range(1, 31))]
    fig = plt.figure(figsize=[15, 5])
    ax = fig.add_subplot(111)
    ax.plot(df[cols].loc[index, :])
    ax.set_title(model + ' '+ metric + unit, fontsize=fontsize)
    ax.set_ylabel(metric, fontsize=fontsize)
    ax.set_xlabel("trials", fontsize=fontsize)
    
```

```python
model = 'svc_123'
metric_1 = 'Billed Duration'
metric_2 = 'Max Memory Used'  
metric_plot(model, metric_1, metric_2)


# metric_plot(model, metric)

metric = 'execution_times'  
execution_time_plot(model)

time_series_plot('svc_123', 'Max Memory Used', 2)
time_series_plot('svc_123', 'Billed Duration', 2)
```

```python
def plot_index_cost(index, allocated_memory):
    df = pd.read_csv('/Users/michaelzhang/Downloads/Seneca/results/spreadsheets/svc_123_Billed Duration.csv')
    df_2 = pd.read_csv('/Users/michaelzhang/Downloads/Seneca/results/spreadsheets/svc_123_Max Memory Used.csv')
    cols = [str(x) for x in list(range(1, 31))]
    cost_list = df[cols].loc[index]/1000 * allocated_memory/1024 * 0.00001667
    memory_used = df_2[cols].loc[index]
    fig = plt.figure(figsize=[20, 10])
    ax = fig.add_subplot(211)
    ax.set_title("Lambda cost", fontsize=fontsize)
    ax.set_xlabel("Trials", fontsize=fontsize)
    ax.set_ylabel("Cost ($)", fontsize=fontsize)
    ax.plot(cost_list)
    ax = fig.add_subplot(212)
    ax.set_title("Max Memory Used", fontsize=fontsize)
    ax.set_xlabel("Trials", fontsize=fontsize)
    ax.set_ylabel("Max Memory Used (MB)", fontsize=fontsize)
    ax.plot(memory_used)
    
```

```python
def plot_minmax(allocated_memory, left, right):
    df = pd.read_csv('/Users/michaelzhang/Downloads/Seneca/results/spreadsheets/svc_123_Billed Duration.csv')
    cols = [str(x) for x in list(range(1, 31))]
    max_cost = df[cols].max(axis = 1).loc[left:right]
    min_cost = df[cols].min(axis = 1).loc[left:right]
    fig = plt.figure(figsize=[30, 10])
    ax = fig.add_subplot(111)
    ax.set_title("Max & Min Billed Duration", fontsize=fontsize)
    ax.set_ylabel("Billed Duration (ms)", fontsize = fontsize)
    ax.set_xlabel("Trials", fontsize=fontsize)
    ax.plot(max_cost)
    ax.plot(min_cost)
```

```python
plot_minmax(823, 1, 100)
```

```python
def plot_total_cost(model, allocated_memory):
    df = pd.read_csv('/Users/michaelzhang/Downloads/Seneca/results/spreadsheets/{0}_Billed Duration.csv'.format(model))
    cols = [str(x) for x in list(range(1, 31))]
    cost_sum = df[cols].sum()/1000 * allocated_memory/1024 * 0.00001667
    print ("Average total cost: {0}".format(sum(cost_sum) / len(cost_sum)))
    print ("Highest cost: {0}".format(max(cost_sum)))
    print ("Lowest cost: {0}".format(min(cost_sum)))
    print ("Cost Difference between high and low: {0}".format(max(cost_sum) - min(cost_sum)))
    fig = plt.figure(figsize=[30, 10])
    ax = fig.add_subplot(111)
    ax.set_title(model + " Total Cost at {0} (MB)".format(allocated_memory), fontsize=fontsize)
    ax.set_ylabel("Cost ($)", fontsize = fontsize)
    ax.set_xlabel("Trials", fontsize=fontsize)
    ax.plot(cost_sum)
```
